```python
import os

# list out all files in the DataSet folder.
data_files = os.listdir('/Users/violacitra/Data Analysis/Cyclistic Trip Data Set 2020 - 2024/DataSet')
data_files
```

```python
# yield df helps us to 'generate' table instead of returning or printing them. 
# Then, we're gonna use concat() function to append the row after we generate another table.
def load_files(data_files):
    for filename in data_files:
        # Read data from each file
        df = pd.read_csv(f'/Users/violacitra/Data Analysis/Cyclistic Trip Data Set 2020 - 2024/DataSet/{filename}')
        yield df 
```

```python
# Generate a list of file names
data = [x for x in data_files]


# load_files takes 1 argument (a list of file names)
CyclisticData = pd.concat(load_files(data))

# Reset index
CyclisticData.reset_index(drop=True, inplace=True)

# Print
CyclisticData
```

```python
# saving the combined CSVs
CyclisticData.to_csv('CombinedCyclisticData.csv')
```
