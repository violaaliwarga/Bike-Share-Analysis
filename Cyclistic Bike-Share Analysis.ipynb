# 3.2. Combine All CSV Files
There are 52 .csv files that's needed to be processed

```python
import os
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# list out all files in the DataSet folder.
data_files = os.listdir('/Users/violacitra/Data Analysis/Cyclistic Trip Data Set 2020 - 2024/DataSet')
data_files
```

```python
def load_files(data_files):
    for filename in data_files:
        # Read data from each file
        df = pd.read_csv(f'/Users/violacitra/Data Analysis/Cyclistic Trip Data Set 2020 - 2024/DataSet/{filename}')
        # yield df helps us to 'generate' table instead of returning or printing them. 
        yield df 
```

```python
# Generate a list of file names
data = [x for x in data_files]


# load_files takes 1 argument (a list of file names)
# concat() to append the row after we generate another table.
CyclisticData = pd.concat(load_files(data))

# Reset index
CyclisticData.reset_index(drop=True, inplace=True)

# Print
CyclisticData
```

```python
CyclisticData.info()
```

# 3.3. Handling Missing Data

### 3.3.1. Check for Missing Values

```python
df = CyclisticData

missing_data_count = df.isnull().sum()
total_missing = missing_data_count.sum()
total_rows = len(df)
missing_data_percentage = (missing_data_count / total_rows) * 100
total_percentage_missing = (total_missing / total_rows) * 100

print(missing_data_count[missing_data_count > 0])
print(missing_data_percentage[missing_data_percentage > 0])
print(f"Percentage of missing data: {total_percentage_missing:.2f}%")
```

### Identify Missing Values (Flag)
Since the missing values % is quite significant, I would like to examine them more by flagging them.

```python
# Use .astype(int) to change False & True to boolean 0 & 1 to make the next steps easier

df_flag = df

df_flag['start_station_name_flag'] = df_flag['start_station_name'].isnull().astype(int)
df_flag['start_station_id_flag'] = df_flag['start_station_id'].isnull().astype(int)
df_flag['end_station_name_flag'] = df_flag['end_station_name'].isnull().astype(int)
df_flag['end_station_id_flag'] = df_flag['end_station_id'].isnull().astype(int)
df_flag['end_lat_flag'] = df_flag['end_lat'].isnull().astype(int)
df_flag['end_lng_flag'] = df_flag['end_lng'].isnull().astype(int)
```

### Plot the Distribution Values of the Flagged df

```python
df_flag[['start_station_name_flag', 'start_station_id_flag', 'end_station_name_flag', 'end_station_id_flag',
    'end_lat_flag', 'end_lng_flag']].sum().plot(kind='bar')
plt.title('Count of Missing Values by Column')
plt.xlabel('Columns')
plt.ylabel('Count of Missing Values')
plt.show()
```

### Correlation Analysis of Missing Data Flags
To analyze how the missing values in one column correlate with missing values in the other

```python
# Compute the correlation matrix of missing value flags
missing_flags = df_flag[['start_station_name_flag', 'start_station_id_flag', 'end_station_name_flag',
                         'end_station_id_flag', 'end_lat_flag', 'end_lng_flag']]
missing_corr = missing_flags.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(missing_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Missing Value Flags Across Columns')
plt.show()
```

### Handling the Missing Data

```python
df_drop2 = df  # Load your data file

# Drop flagged columns
df_drop2_cleaned = df_drop2.drop(columns=['start_station_name_flag', 'start_station_id_flag',
                                         'end_station_name_flag', 'end_station_id_flag', 'end_lat_flag',
                                         'end_lng_flag'])

df_drop2_cleaned.info()
```

```python
# Drop rows where 'end_lat' or 'end_lng' have missing values
df_cleaned = df_drop2_cleaned.dropna(subset=['end_lat', 'end_lng'])

# Display the first few rows and the data info to verify
print(df_cleaned.head())
df_cleaned.info()
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler

df_KNN = df_cleaned.copy()

# 1. Preprocess the Data
# List columns to exclude from scaling/imputation
exclude_columns = ['ride_id']
                   
#'rideable_type', 'started_at', 'ended_at']

# Convert categorical columns to numerical using LabelEncoder (excluding `ride_id`)
label_encoders = {}
categorical_columns = [col for col in df_KNN.columns if df_KNN[col].dtype == 'object' and
                       col not in exclude_columns]

for col in categorical_columns:
    label_encoders[col] = LabelEncoder()
    df_KNN[col] = label_encoders[col].fit_transform(df_KNN[col].astype(str))

# Select only the columns that should be imputed
columns_to_impute = df_KNN.columns.difference(exclude_columns)

# 2. Apply KNN Imputation
imputer = KNNImputer(n_neighbors=5)

# Normalize the data
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_KNN[columns_to_impute]), columns=columns_to_impute, index=df_KNN.index)

# Perform KNN Imputation
df_imputed_scaled = pd.DataFrame(imputer.fit_transform(df_scaled), columns=columns_to_impute, index=df_KNN.index)

# Inverse transform the scaling
df_imputed = pd.DataFrame(scaler.inverse_transform(df_imputed_scaled), columns=columns_to_impute, index=df_KNN.index)

# 3. Post-process the Data
# Convert numerical columns back to original categorical where needed
for col in categorical_columns:
    df_imputed[col] = label_encoders[col].inverse_transform(df_imputed[col].round().astype(int))

# Reattach the excluded columns (ensuring alignment by using the index)
df_final = pd.concat([df_KNN[exclude_columns], df_imputed], axis=1)

# Display the imputed DataFrame
print(df_final.head())
```

### Assigning Appropriate Type to started_at & ended_at

```python
cleaned_df = df_final.copy()

# Strip any trailing fractional seconds or extra characters, there's some in the data
cleaned_df['started_at'] = cleaned_df['started_at'].str.split('.', n=1).str[0]
cleaned_df['ended_at'] = cleaned_df['ended_at'].str.split('.', n=1).str[0]

# Converting to datetime
cleaned_df['started_at'] = pd.to_datetime(cleaned_df['started_at'])
cleaned_df['ended_at'] = pd.to_datetime(cleaned_df['ended_at'])
```

```python
cleaned_df.info()
cleaned_df.isnull().sum()
```

# Creating New Columns Necessary for Analysis

```python
# Extract the month from 'started_at' and convert to month names
cleaned_df['month'] = cleaned_df['started_at'].dt.month_name()

# Extract the day of the week from 'started_at'
cleaned_df['day'] = cleaned_df['started_at'].dt.day_name()

# Extract the hour from 'started_at'
cleaned_df['hour'] = cleaned_df['started_at'].dt.hour

# Calculate the duration in minutes
cleaned_df['duration_minutes'] = round((cleaned_df['ended_at'] - cleaned_df['started_at']).dt.total_seconds() / 60, 2)

# Display first few rows & data info to verify
print(cleaned_df.head())
cleaned_df.info()
```
